```python
# Load dataset (replace filename if needed)
df = pd.read_csv(data_path("diabetes.csv"))
print("Rows:", len(df))
df.head()
```

```python
# Persistent Database Connection (project-level)
db_path = Path("../data/diabetes.db")
conn = sqlite3.connect(db_path)
print("✅ Connected to persistent database", db_path)

# Optional: confirm tables
pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)
```

```python
# In-memory SQL engine for quick queries
con = sqlite3.connect(":memory:")
df.to_sql("diabetes", con, index=False, if_exists="replace")

# Example query
pd.read_sql_query("SELECT COUNT(*) AS n_rows FROM diabetes;", con)
```

```python
# make sure /reports/sql exists
sql_dir = report_path("sql")
sql_dir.mkdir(parents=True, exist_ok=True)
```

```python
rows = pd.read_sql_query("SELECT COUNT(*) AS n_rows FROM diabetes;", con)
outcomes = pd.read_sql_query(
    "SELECT Outcome, COUNT(*) AS n FROM diabetes GROUP BY Outcome ORDER BY n DESC;", con
)
rows.to_csv(sql_dir / "rows_count.csv", index=False)
outcomes.to_csv(sql_dir / "outcome_counts.csv", index=False)
rows, outcomes
```
