```python
# Path safety: ensures all reads/writes go to top-level /data and /reports
import sys
from pathlib import Path
sys.path.append(str(Path.cwd().resolve().parents[0]))  # parent of /notebooks
from paths import data_path, report_path, print_layout
print_layout()

# Common imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sqlite3
```

```python
# Ensure common output subfolders exist
for sub in ["artifacts", "sql", "eda", "models"]:
    d = report_path(sub)
    d.mkdir(parents=True, exist_ok=True)
```

```python
# Load dataset (replace filename if needed)
df = pd.read_csv(data_path("diabetes.csv"))
print("Rows:", len(df))
df.head()
```

```python
# Persistent Database Connection (project-level)
db_path = Path("../data/diabetes.db")
conn = sqlite3.connect(db_path)
print("âœ… Connected to persistent database", db_path)

# Optional: confirm tables
pd.read_sql_query("SELECT name FROM sqlite_master WHERE type='table';", conn)
```

```python
# In-memory SQL engine for quick queries
con = sqlite3.connect(":memory:")
df.to_sql("diabetes", con, index=False, if_exists="replace")

# Example query
pd.read_sql_query("SELECT COUNT(*) AS n_rows FROM diabetes;", con)
```

```python
# Example: Save sample output to /reports/
df.head(10).to_csv(report_path("artifacts/sample_head10.csv"), index=False)
print("Saved ->", report_path("artifacts/sample_head10.csv"))
```

```python
# Cleanup
conn.close()
con.close()
print("ðŸ”’ All connections closed.")
```
